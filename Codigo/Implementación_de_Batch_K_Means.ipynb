```python
import pandas as pd
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
df_accidentes = pd.read_csv("X.csv")
```

# Aprendizaje no supervisado

```python
kmeans = KMeans(n_clusters=2).fit(df_accidentes)
print(f'Total de centros {len(kmeans.cluster_centers_)}')
print("="*34)
print(f'Arreglo con la longitud {kmeans.predict(df_accidentes)}')
```

```python
df_accidentes['group'] = kmeans.predict(df_accidentes)

df_accidentes.head(10)
```

# Visualizacion de la agrupación de los cluster

```python
sns.pairplot(df_accidentes, palette="husl" , hue='group', kind="reg" )
```

```python
sns.pairplot(df_accidentes[['DIA','PERIODO','DIA_SEMANA','MES','NUM_COMUNA','NUM_GRAVEDAD','NUM_DISENO','NUM_CLASE','group']],kind="reg" , hue = 'group',  palette="husl")
```
