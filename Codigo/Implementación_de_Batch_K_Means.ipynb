```python
import pandas as pd
import seaborn as sns

from sklearn.cluster import MiniBatchKMeans
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
df_accidentes = pd.read_csv("X.csv")
```

# Aprendizaje no supervisado

```python
kmeans = MiniBatchKMeans(n_clusters=2, batch_size=8).fit(df_accidentes)
print(f'Total de centros {len(kmeans.cluster_centers_)}')
print("="*34)
print(f'Arreglo con la longitud {kmeans.predict(df_accidentes)}')
```

```python
df_accidentes['group'] = kmeans.predict(df_accidentes)

df_accidentes.head(10)
```

```python
df_accidentes.to_csv("K-Means.csv", index = False)
```
